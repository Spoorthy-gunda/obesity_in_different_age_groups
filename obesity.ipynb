1. LocationAbbr and LocationID can be deleted as we have locationDesc
2. Similarly, ClassID, TopicID, QuestionID, StratificationCategoryId1,StratificationID1 columns can be deleted.
3. Data value Unit has no values so we can ignore the column, similarly DataValueTypeID is the same for everything.
4. Data_value_Alt and the Data_value are the same, in the description it is mentioned that Data_value_Alt is the numerical representation of Data_value but both of them have float as the data type.
5. Datasource is the same for every data point so we can ignore that column.
6. GeoLocation has the lattitude and longitude values of the location, so we can ignore that column as it does not provide any other extra information about te location apart from locationID. 


```python
data =  data.drop(columns=['LocationAbbr', 'Datasource', 'Class', 'High_Confidence_Limit ', 'Low_Confidence_Limit', 'Topic', 'Data_Value_Unit', 'Data_Value', 'Data_Value_Type','GeoLocation', 'DataValueTypeID', 'StratificationCategory1', 'Stratification1'])
data.info()
```

#### Observation: 
Data_Value_Footnote_Symbol or the Data_value_Footnote is present when the Data_Value_Alt is not available in that row. 53392 is the total rows, foot note is present for 5046 rows and data values ae available for 48346(48346+5046 = 53392).

```python
data1 = (data['Data_Value_Footnote'].isnull()|data['Data_Value_Footnote_Symbol'].isnull())
data = data[data1]
data = data.drop(columns = ['Data_Value_Footnote_Symbol', 'Data_Value_Footnote'])
data.info()
```

```python
for col in data:    
    print(col, data[col].nunique())
```

```python
data = data.astype({'Gender':'category', 'ClassID':'category', 'TopicID': 'category', 'QuestionID': 'category', 'LocationID':'category', 'StratificationCategoryId1':'category', 'StratificationID1': 'category', 'Gender': 'category', 'Race/Ethnicity': 'category', 'Education': 'category'})
data.info()
```

#### Observation:
There are 6 stratificationCategoryId's and they correspond to Age, Education, Gender, Race/Ethnicity, Income, Total. Only the value corresponding to the StratificationCategory is present. So we can divide the data into 6 different sets, and perform individual analysis.

```python
data_Age = data['Age(years)'].notnull()
dataAge= data[data_Age]
dataAge = dataAge.drop(columns = ['Total', 'Education', 'Gender', 'Income', 'Race/Ethnicity'])
dataAge.info()
```

Lets now check if the number of datapoints for each category is balanced or not.

```python
print(dataAge.groupby(['QuestionID', 'Age(years)']).count()['Sample_Size'])
```

```python
dataAgeQ1 = dataAge.query('Question == "Percent of adults aged 18 years and older who have obesity"')
sns.catplot(x="Age(years)", y="Data_Value_Alt", data=dataAgeQ1, aspect =2.5)
```

```python
dataAgeQ2 = dataAge.query('Question == "Percent of adults aged 18 years and older who have an overweight classification"')
print(dataAgeQ2.groupby('Age(years)').count()['Sample_Size'])
sns.catplot(x="Age(years)", y="Data_Value_Alt", data=dataAgeQ2)
```

```python
sns.catplot(x="QuestionID", y="Data_Value_Alt", data=dataAge, aspect =2.5, hue = 'Age(years)', kind ='bar')
```

1. Q036, Q037 corresponds to the percent of adults who have obesity and overweight respectively. It is clear that the percent of adults aged between 18-24 who are obese/overweight is less when compared to the other aged groups. 
2. The percent of adults who are overweight is increasing with the age, but the percent of adults who are obese is more for the age group 45-54.
3. Overall more percent of adults aged between 18-24 eat at least one fruit/vegetable (Q018 and Q019) and get more excercise(Q044, Q045, Q046) when compared to other aged groups and percent of adults who engege in no leisure time physical activity is less in the aged group 18-24.

```python
data_Gen = data['Gender'].notnull()
dataGen= data[data_Gen]
dataGen = dataGen.drop(columns = ['Total', 'Age(years)', 'Education', 'Income', 'Race/Ethnicity'])
dataGen.info()
```

Percentage of males who eat at least one fruit/vegetable per day is greater than the percentage of females who eat at least one fruit/vegetable per day, similarly percentage of males who get excercise is greater than the percentage of females. But the percentage of males who are overweight is a lot more than the percentage of females who are overweight. The reason for this is probably becuase the physiological differences between male and female.
